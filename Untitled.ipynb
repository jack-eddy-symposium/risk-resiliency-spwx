```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
np.random.seed(42)

num_samples = 5000
X1 = np.random.gamma(shape=2, scale=3, size=num_samples)  # Wildfires intensity
```

```python

# Create density plot
plt.figure(figsize=(8, 5))
sns.kdeplot(X1, fill=True, color="blue", alpha=0.5)
plt.xlabel("Value")
plt.ylabel("Density")
plt.title("Density Plot of Gamma Distribution")
plt.grid(True)

# Show plot
plt.show()
```
