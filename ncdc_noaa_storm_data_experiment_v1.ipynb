```python
# https://www.ncdc.noaa.gov/swdiws

ncdc_base_url = "https://www.ncdc.noaa.gov/swdiws"

# ncdc_dataset = 'nx3structure:inv'
ncdc_dataset = 'nx3structure'
ncdc_output_format = 'json'
ncdc_daterange = '20230331:20230401'
url = f"{ncdc_base_url}/{ncdc_output_format}/{ncdc_dataset}/{ncdc_daterange}"
```

```python
# Make the request
response = requests.get(url)

# Check response status
if response.status_code == 200:
    ncdc_data = response.json()
    print("Data retrieved successfully!")
    print(ncdc_data)  # Print or process data as needed
else:
    print(f"Failed to retrieve data: {response.status_code}")
```

```python
# Convert storm data into a DataFrame
# # Ensure all numeric values are converted properly
# for entry in ncdc_data:
#     entry['MAX_REFLECT'] = float(entry['MAX_REFLECT'])
#     entry['VIL'] = float(entry['VIL'])
#     entry['AZIMUTH'] = float(entry['AZIMUTH'])
#     entry['RANGE'] = float(entry['RANGE'])
ncdc_df = pd.DataFrame(ncdc_data['result'])

# Extract Longitude and Latitude from the SHAPE field
ncdc_df[['Longitude', 'Latitude']] = ncdc_df['SHAPE'].str.extract(r'POINT \((-?\d+\.\d+) (-?\d+\.\d+)\)').astype(float)


# Choose a ZTIME to visualize (Change this to filter different timestamps)
selected_ztime = "2023-03-31T19:38:53Z"
filtered_ncdc_df = ncdc_df[ncdc_df['ZTIME'] == selected_ztime]

# Create a Folium map centered on the US
m = folium.Map(location=[39.8283, -98.5795], zoom_start=4)  # Centered in the US

# Check if there is data for the selected ZTIME
if not filtered_ncdc_df.empty:
    # Add storm data points to the map
    for _, row in filtered_ncdc_df.iterrows():
        popup_info = (f"ZTIME: {row['ZTIME']}<br>"
                      f"MAX_REFLECT: {row['MAX_REFLECT']}<br>"
                      f"VIL: {row['VIL']}<br>"
                      f"AZIMUTH: {row['AZIMUTH']}<br>"
                      f"RANGE: {row['RANGE']}")

        folium.CircleMarker(
            location=[row['Latitude'], row['Longitude']],
            radius=int(row['MAX_REFLECT']) / 10,  # Scale marker size by MAX_REFLECT
            color='red',
            fill=True,
            fill_color='orange',
            fill_opacity=0.7,
            popup=folium.Popup(popup_info, max_width=300)
        ).add_to(m)

    print(f"Displaying storm data for {selected_ztime}")
else:
    print(f"No storm data available for {selected_ztime}")

# Save map to an HTML file and display it
# m.save("storm_map_filtered.html")
m
```
